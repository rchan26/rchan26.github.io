## Recap of [Part I](https://github.com/alan-turing-institute/DataScienceSkills/blob/bayes_intro/bayesian_inference/part_i.ipynb)

In Part I, we saw that in Bayesian modelling, we are interested in performing inference for a *posterior distribution* over all our parameters $\boldsymbol{\theta}$ given some data $\boldsymbol{x}$, as opposed to computing a single point estimate. This has the advantage of providing us a nice way to talk about *uncertainty*.

The *posterior distribution* of our parameters has probability density

$$
\color{purple}{\pi(\boldsymbol{\theta}|\boldsymbol{x})} = \frac{\color{blue}{l(\boldsymbol{x}|\boldsymbol{\theta})}\color{red}{\pi(\boldsymbol{\theta})}}{\color{green}{\pi(\boldsymbol{x})}},
$$
where 
* <font color='blue'>$l(\boldsymbol{x}|\boldsymbol{\theta})$ is the *likelihood* function</font>,
    * which gives us the probability of observing our data $\boldsymbol{x}$ given a choice of parameters $\boldsymbol{\theta}$
* <font color='red'>$\pi(\boldsymbol{\theta})$ is the *prior*</font>,
    * which encodes our beliefs about what values our parameters might take *before* seeing the data $\boldsymbol{x}$
* <font color='green'>$\pi(\boldsymbol{x}):=\int_{\Theta}l(\boldsymbol{x}|\boldsymbol{\theta})\pi(\boldsymbol{\theta})d\boldsymbol{\theta}$ is the *normalising constant* (also referred to as the *evidence*, or *marginal likelihood*)</font>
    * which ensures that the density integrates (or sums) to 1 (if a density doesn't integrate/sum to 1, then it is not a probability distribution)

We saw in Part I, that a key difficulty in performing Bayesian inference is that to compute the posterior, we must compute intractable integrals. In most practical problems, computing the integral on the denominator (or sum if we have distribution over discrete state space) analytically is going to be a nightmare (or even intractable).

Most approximating techniques therefore focus on performing inference without having to compute the <font color='green'>evidence</font>.

**NB:** In fact, many challenges in statistics boils down to computing some integral. For example:
1. **Computing expectations:** We often want to obtain *summary statistics* of the form:

$$
\mathbb{E}_{\pi(\boldsymbol{\theta}|\boldsymbol{x})}[f(\boldsymbol{\theta})] = \int_{\Theta} f(\boldsymbol{\theta})\pi(\boldsymbol{\theta}|\boldsymbol{x}) d\boldsymbol{\theta}.
$$

2. **Computing the *posterior predictive distribution*** (for forecasting):

$$
\pi(\boldsymbol{x}^{*}|\boldsymbol{x}) = \int l(\boldsymbol{x}^{*}|\boldsymbol{\theta}) \pi(\boldsymbol{\theta}|\boldsymbol{x}) d\boldsymbol{\theta}
$$

This denotes the probability of *unseen* data $\boldsymbol{x}^{*}$ conditional on the inference that has been performed after the data $\boldsymbol{x}$ has been observed.

3. **Marginalisation:** Given the joint posterior of $(\theta_{1}, \theta_{2}) \in (\Theta_{1}, \Theta_{2})$, we may often be intersted in the *marginal posterior*:

$$
\pi(\theta_{1}|\boldsymbol{x}) = \int_{\Theta_{2}} \pi(\theta_{1},\theta_{2}|\boldsymbol{x}) d\theta_{2}.
$$

4. **Cumulative probabilities:** If $\theta\in\mathbb{R}$ and we wanted to answer "what is the probability that $\theta$ lies in [-0.5, 1] given that we've $\boldsymbol{x}$?", this is given by:

$$
P\left(\theta \in [-0.5, 0.1]|\boldsymbol{x}\right) = \int_{-0.5}^{1} \pi(\theta|\boldsymbol{x}) d\theta.
$$

## Bayesian Inference in practice

Restricting your modelling to a small subset of models which can be analysed analytically is of course going to limit your ability to model a problem well, and so we would like to develop methodology to approximate our posterior.

At the end of Part I we looked at using a *grid approximation* (i.e. evaluating the un-normalised posterior at several points on a grid and then normalising), which has several drawbacks:
* **Dimensionality scaling:** Adding a single parameter to our model adds a dimension to our grid
* **Parameter ranges:** In practice, how do we choose what range of values to evaluate at? How do we know we aren't missing interesting areas of the space?
* **Usefulness of output:** We're typically interested in quantities *related* to the posterior more so than the actual value of the density at some points
    * e.g. it's not entirely straightforward to estimate something like "what is the probability that $\theta$ lies in [-0.5, 1] given our data?" using the grid approximation output

In this talk, we look at *Monte Carlo methods* and in particular, we will focus on *Markov chain Monte Carlo* and its application in performing inference Bayesian inference (i.e. fitting Bayesian models).

## Monte Carlo approximations

A *Monte Carlo* method is a technique which uses *random* simulation and probability to solve complex problems. 

In our context, the key idea of Monte Carlo simulation is that if we can draw a N *independent, identically distributed (i.i.d.)* samples $\boldsymbol{\theta}_{i} \sim \pi(\boldsymbol{\theta}|\boldsymbol{x})$ for $i=1,\dots,N$, from the posterior, then we can unbiasedly estimate expectations with the *Monte Carlo estimate*:

$$
\lim_{N\rightarrow\infty} \frac{1}{N}\sum_{i=1}^{N}f(\boldsymbol{\theta}_{i}) = \int_{\Theta} f(\boldsymbol{\theta})\pi(\boldsymbol{\theta}|\boldsymbol{x}) d\boldsymbol{\theta}.
$$

So by being able to sample from the posterior distribution, we are subsequently able to approximate the density *and* any expectations with respect to it (for example the integrals above). Critically, the Monte Carlo approximation converges to this quantity as we increase the same size $N$.

Keep in mind that the reason why we want to sample from the posterior is because it allows us to approximate the distribution *and* compute quantities respect to it (e.g. use it to make predictions).

In essensence, Monte Carlo provides a way for us to perform inference with our model/posterior which we could not have done otherwise analytically!

**NB:** In this talk, we focus on the application of Monte Carlo and Markov chain Monte Carlo to sample from our Bayesian posterior distribution, but to clarify, they can be used to sample from a general probability distribution.

**NB:** A bit more intuition of Monte Carlo approximations of distributions can be found in Section 1 of [Part II - Supplementary Material](https://rchan26.github.io/pdfs/talks/bayesian_inference_talk_part_ii/part_ii_supplementary.ipynb).

## Monte Carlo methods

While Monte Carlo provides an elegant algorithm to estimate integrals, it is not entirely clear how we could actually simulate i.i.d. samples from $\pi(\boldsymbol{\theta}|\boldsymbol{x})$. This is typically the key complication with Monte Carlo methods, and there are plethora of methods:

**NB:** MCMC don't actually return i.i.d. samples fom $\pi$ but the methods still allow you to approximate the distribution and expectations with respect to it. We will see how this affects how we use the output from MCMC later on...

*Markov chain Monte Carlo (MCMC)* is one of the most widely used methods for drawing samples from posterior distributions and critically *does not* require that you can compute the normalising constant. In particular, MCMC algorithms are methods for generating samples $\boldsymbol{\theta}_{i}$ while exploring the parameter space $\Theta$ using a *Markov chain* mechanism. This mechanism is constructed so that the samples $\boldsymbol{\theta}_{i}$ mimic samples drawn from the posterior $\pi(\boldsymbol{\theta}|\boldsymbol{x})$.

We use MCMC when we cannot evaluate from $\pi(\boldsymbol{\theta}|\boldsymbol{x})$ directly, but can evaluate $\pi(\boldsymbol{\theta}|\boldsymbol{x})$ up to a normalising constant, i.e. we can compute the *un-normalised posterior density*:
$$
\pi(\boldsymbol{\theta}|\boldsymbol{x}) \propto l(\boldsymbol{x}|\boldsymbol{\theta})\pi(\boldsymbol{\theta}).
$$

We will discuss the Metropolis-Hastings (MH) algorithm which is the simplest and most popular MCMC method. In fact, most practical MCMC algorithms can be seen as special cases or extensions of MH. But first, what is a Markov chain?!

## Markov chains

The sampling procedure of a MCMC algorithm is that of a *Markov chain*, which is Markov process with a discrete state space _or_ a discrete index set which means the next step in the process depends *only* on the present location of the chain, and importantly *does not* depend on any other past history. 

There is a **lot** of theory on Markov processes, but for our purposes in this notebook where our aim is to focus on the *practical* rather than the theoretical, we will only discuss these very briefly and concentrate on *discrete-time Markov chains* - but for those interested, there is a bit more on Markov chains (on discrete spaces) in [Part II - Supplementary Material](https://rchan26.github.io/pdfs/talks/bayesian_inference_talk_part_ii/part_ii_supplementary.ipynb).

**NB:** *state space* means the possible values that the process can take, the *index set* often represents time but it doesn't necessarily have to.

A *discrete-time Markov chain* is sequence of random variables $X_{1}, X_{2}, \dots$, with the *Markov property* (or *memory-less property*).

$$
p(X_{n}=x_{n} | X_{n-1}=x_{n-1}, \dots, X_{0}=x_{0}) = p(X_{n}=x_{n} | X_{n-1}=x_{n-1}),
$$

where $p$ is the *transition density* of the Markov chain.

### Stationary distribution of a Markov chain

The *stationary distribution* (or *invariant distribution*) of a Markov chain describes the distribution of $X_{t}$ after a 'sufficiently long' time that the distribution of $X_{t}$ no longer changes - i.e. it's reached some sort of *steady state*.

**NB:** more intuition about stationary distributions can be found in Section 2 of [Part II - Supplementary Material](https://rchan26.github.io/pdfs/talks/bayesian_inference_talk_part_ii/part_ii_supplementary.ipynb).

## Markov chain Monte Carlo

The key fundamental property of the Markov chain which is constructed in MCMC algorithms is that it converges to a *stationary distribution* (or *invariant distribution*) which coincides with our target posterior distribution. This means that once it reaches the stationary distribution, we can continue our chain and take these as samples from our posterior distribution.

For any starting point, the chain will converge to some stationary distribution $\pi(x)$ as long as it obeys the following properties:
1. *Irreducibility*: for any state of the Markov chain, there is a *positive* probability of visiting all other states.
2. *Aperiodicity*: the chain should not be able to get *trapped* in any cycles (we do not want to create an process which can get stuck in part of the state space).

**NB:** Because of this Markov chain construction, MCMC samplers do not actually generate i.i.d. draws from the posterior distribution. The samples it generates are correlated (because they depend on the previous state of the chain). Furthermore, the theory here suggests that we can initialise our chain anywhere and our chain should converge to the stationary distribution (which is our target posterior distribution in this case). But this means that the samples generated from the whole chain aren't from the target distribution. In practice, we need to 'cut-off' some initial samples which is called the *burn-in* period. We will look at the implications of this on MCMC algorithms later on.

**NB:** MCMC samplers which satisfy detailed balance are known as *reversible MCMC algorithms* but note that there are some recent developments within the literature to construct chians which are *non-reversible* (which are meant to have some nice properties but this is way outside the scope for this talk).

## Metropolis-Hastings

To recap, we can use MCMC to generate samples from our target posterior distribution and we just require that we can evaluate our target density up to a normalising constant:
$$
\pi(\boldsymbol{\theta}|\boldsymbol{x}) \propto l(\boldsymbol{x}|\boldsymbol{\theta})\pi(\boldsymbol{\theta}),
$$

*and* it requires us to choose some *proposal distribution* $q(\boldsymbol{\theta}^{*}|\boldsymbol{\theta})$.

The **Metropolis-Hastings (MH) algorithm** (a more general version of the original Metropolis algorithm) is as follows:

1. Initialise $\boldsymbol{\theta}_{0}$.
2. For $i=1,\dots,N$:

    i. Sample $u\sim\mathcal{U}(0,1)$.
    
    ii. Sample $\boldsymbol{\theta}^{*} \sim q(\boldsymbol{\theta}^{*}|\boldsymbol{\theta}_{i-1})$.

    iii. Compute *acceptance probability* 
    $$
    \alpha(\boldsymbol{\theta}_{i-1}, \boldsymbol{\theta}^{*}) := \min \left\{ 1, \frac{\pi(\boldsymbol{\theta}^{*}|\boldsymbol{x})q(\boldsymbol{\theta}_{i-1}|\boldsymbol{\theta}^{*})}{\pi(\boldsymbol{\theta}_{i-1}|\boldsymbol{x})q(\boldsymbol{\theta}^{*}|\boldsymbol{\theta}_{i-1})} \right\}.
    $$

    iv. If $u \leq \alpha(\boldsymbol{\theta}_{i-1}, \boldsymbol{\theta}^{*})$, then set $\boldsymbol{\theta}_{i} = \boldsymbol{\theta}^{*}$.
    
    v. Otherwise, reject and set $\boldsymbol{\theta}_{i} = \boldsymbol{\theta}_{i-1}$. 

The MH algorithm is very simple, but it requires careful design of the proposal distribution $q(\boldsymbol{\theta}^{*}|\boldsymbol{\theta})$. 

The key take away from this is that the algorithm constructs a Markov chain whose stationary distribution coincides with the posterior distribution of interest. We won't prove this here, but we cover very briefly some intution for this here.

#### Why does this work?

MCMC samplers construct irreducible and aperiodic Markov chains which have the target distribution as the stationary distribution. A *sufficient* but not neccessary condition to ensure that the stationary distribution is given by a particular density $\pi(x)$ is the *detailed balance* (or *reversibility*) condition:

$$
\pi(x_{i})p(x_{i-1}|x_{i}) = \pi(x_{i-1})p(x_{i}|x_{i-1}).
$$

This simply means that the probability of being at $x_{i}$ and moving to $x_{i-1}$ is the same as the probability of being at $x_{i-1}$ and moving to $x_{i}$.

Therefore, one way to design MCMC samplers is to ensure that detailed balance is satisfied. However, it is also important that the sampler converges *quickly* to the stationary distribution and most efforts in developing 'better' algorithms are devoted to increasing this convergence speed (although there are other areas where people are looking to improve MCMC algorithms). 

In theory, under some conditions (like ensuring the proposal would allow you to explore the complete space - we can do this by making sure $q(\boldsymbol{x}^{*}|\boldsymbol{x}_{i-1})$ is positive for all $\boldsymbol{x}^{*}$ and $\boldsymbol{x}_{i-1}$ in the support for $\pi$ as this means there is a non-zero probability of reaching any point in the space with zero probability), we should be able to start our chain anywhere and our Metropolis-Hastings sampler should be able to converge to the posterior distribution and we can obtain samples from our posterior distribution.

Spoiler: in practice, it is a bit more complicated than this...

### Independent sampler

If we choose $q(\boldsymbol{\theta}^{*}|\boldsymbol{\theta}_{i-1})=q(\boldsymbol{\theta}^{*})$ (i.e. the proposal is _independent_ of the current state), we obtain the *independent sampler* which has acceptance probability:

$$
\alpha(\boldsymbol{\theta}_{i-1}, \boldsymbol{\theta}^{*}) := \min \left\{ 1, \frac{\pi(\boldsymbol{\theta}^{*}|\boldsymbol{x})q(\boldsymbol{\theta}_{i-1})}{\pi(\boldsymbol{\theta}_{i-1}|\boldsymbol{x})q(\boldsymbol{\theta}^{*})} \right\}.
$$

### The Metropolis algorithm

If we choose $q(\boldsymbol{\theta}^{*}|\boldsymbol{\theta}_{i-1})=q(\boldsymbol{\theta}_{i-1}|\boldsymbol{\theta}^{*})$ (i.e. the propopsal is _symmetric_), we obtain the *Metropolis algorithm*, which has acceptance probability:

$$
\alpha(\boldsymbol{\theta}_{i-1}, \boldsymbol{\theta}^{*}) := \min \left\{ 1, \frac{\pi(\boldsymbol{\theta}^{*}|\boldsymbol{x})}{\pi(\boldsymbol{\theta}_{i-1}|\boldsymbol{x})} \right\}.
$$

An important case of the Metropolis algorithm is when the proposal is a Gaussian distribution centred on the current location of the chain. This is also known as the *Random Walk Metropolis Hastings (RWMH) algorithm*.

## Normal distribution example

In this section, we will walk through a simple implementation of the Metropolis-Hastings algorithm to obtain samples from a Normal distribution. In particular, our *target distribution* has density:

$$
\pi(\theta) \propto \exp \left\{-\frac{(\theta-\mu)^{2}}{2\sigma^{2}}\right\},
$$

where $\mu$ and $\sigma^{2}$ denote the *mean* and *variance* of the Normal distribution, respectively.

Here, we will also use a Normal distribution as the proposal distribution centered on the current value (so this corresponds to the Random Walk Metropolis Hastings (RWMH) algorithm). Note that this example is completely trivial - we're sampling from a Normal distribution as our proposal in an MCMC algorithm to sample from a Normal distribution. This is completely just for illustrative purposes, and we will consider a less trivial example later on in this notebook.

Our proposal density is therefore given by:

$$
p(\theta^{*}|\theta) = \frac{1}{\sqrt{2\pi\sigma_{p}^{2}}} \exp \left\{-\frac{(\theta^{*}-\theta)^{2}}{2\sigma_{p}^{2}}\right\},
$$

where $\sigma_{p}^{2}$ is the variance of our proposal which is a *tuning parameter* in our RWMH algorithm.

Here, we implement a Metropolis-Hastings class, where we have to imput a few things at initialisation:
* The initial position of our chain, $\boldsymbol{\theta}_{0}$
* A function to generate a proposal
* A function to evaluate the log proposal density
* A function to evaluate the log posterior / target density
* (Optional): a seed number for our random number generator

Note that we are using *log* proposal and *log* posterior / target densities. The reason why we do this is to avoid computational underflow and overflow (which can occur if we are evaluating the densities in regions of low probability density), and working with logarithms can help us circumvent this issue and is much more stable computationally. As a result, we work with the *log-acceptance probability*:

$$
\log \alpha(\boldsymbol{\theta}_{i-1}, \boldsymbol{\theta}^{*}) := \min \left\{ 0, \log[\pi(\boldsymbol{\theta}^{*}|\boldsymbol{x})q(\boldsymbol{\theta}_{i-1}|\boldsymbol{\theta}^{*})] - \log[\pi(\boldsymbol{\theta}_{i-1}|\boldsymbol{x})q(\boldsymbol{\theta}^{*}|\boldsymbol{\theta}_{i-1})] \right\},
$$

and so at iteration $i=1,\dots,N$, we accept a move to the proposed value $\boldsymbol{\theta}^{*}$ if $\log(u) \leq \log \alpha(\boldsymbol{\theta}_{i-1}, \boldsymbol{\theta}^{*})$ where $u \sim \mathcal{U}(0,1)$.

Now lets create the functions to pass into the function:
* `gen_proposal` is a function which generates a proposal from a Normal distribution centered on a given point
* `log_proposal` is a function which evaluates the log-proposal density at a given point
* `log_posterior` is a function which evaluates the log-target (un-normalised) density at a given point

For this example, we will set $\mu=5$ and $\sigma=0.5$.

```python
# parameters of our target Normal density
mu = 5
sigma = math.sqrt(0.5)

def gen_proposal(sigma):
    return lambda mean: np.random.normal(loc=mean, scale=sigma)

def log_proposal(sigma):
    return lambda x, mean: norm.logpdf(x, loc=mean, scale=sigma)

def log_posterior(mean, sigma):
    return lambda x: -0.5*((x-mean)/sigma)**2
```

Lets just have a look at how this sampler works for a few iterations of an algorithm. Let's start the chain at $\theta_{0}=3$ and let $\sigma_{p}=1$.

```python
%%capture
# initialise the sampler at 3 and let proposal sd = 1
start_point = 3
proposal_sigma = 1
sampler = MetropolisHastings(start_point,
                             gen_proposal(proposal_sigma),
                             log_proposal(proposal_sigma),
                             log_posterior(mu, sigma))

def create_animation(dirname, sampler, iterations):
    for i in range(iterations):
        x = np.linspace(min([mu-4*sigma, sampler.chain[-1]-4*proposal_sigma]),
                        max([mu+4*sigma, sampler.chain[-1]+4*proposal_sigma]), 10000)
        sampler.MH_step_plot(dirname, x)
    
    filenames =[dirname+f"/test_{iteration}.png" for iteration in range(2*(len(sampler.chain)-1)-1)]
    images = []
    for filename in filenames:
        images.append(imageio.imread(filename))
    imageio.mimsave('test.gif', images)
create_animation("raw", sampler, 100);
```

Two choices we need to make for implementing RWMH is:
1. Where do we initialise the chain at?
2. What is the variance of our proposal?

We will look at the impact of these two hyperparameters later, but for a full run, lets now have:
1. $\theta_{0}=-10$
2. $\sigma_{p}^{2}=0.5^{2}$

We let the sampler complete $10000$ iterations...

```python
# initialise the sampler at -10 and let proposal sd = 0.5
start_point = -10
proposal_sigma = 0.5
sampler = MetropolisHastings(start_point,
                             gen_proposal(proposal_sigma),
                             log_proposal(proposal_sigma),
                             log_posterior(mu, sigma))
num_samples = 10000
sampler.MH_sample(num_samples)
```

Lets have a look at how our samples look like by plotting a histogram of our samples and overlay the target density (which is simply a density of $\mathcal{N}(5, 0.5))$.

```python
plt.hist(sampler.chain, bins = 100, density=True)
x = np.linspace(mu-4*sigma, mu+4*sigma, 1000)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.xlabel(r"$\theta$")
plt.ylabel("density")
plt.legend(['True target density', 'Histogram approximation with N={}'.format(num_samples)])
plt.title(r"RWMH with $\sigma_{p}$="+"{} (using full MCMC output)".format(proposal_sigma))
plt.show()
```

So it seems like _most_ of the samples are in the right area but why do we have samples between $-10$ and $0$? If we use these samples to estimate the mean and variance of the target distribution, we have:

```python
print("Mean of MCMC output:", np.mean(sampler.chain))
print("Variance of MCMC output:", np.var(sampler.chain))
```

In our simple example here, we know that our target distribution has mean $5$ and variance $0.5$, but these values are slightly off here. The distribution of our samples from the full MCMC output don't seem to be quite right and there are too many samples towards the left of the distribution.

## Trace plots for MCMC diagnosis

It is useful to consider *trace plots* to diagnose the *convergence* of MCMC plots - these plot the location of the chain over the course of the algorithm.

```python
plt.plot(range(1, len(sampler.chain)+1), sampler.chain)
plt.xlabel("iteration")
plt.ylabel(r"$\theta$")
plt.title(r"Trace plot of RWMH sampler with $\sigma_{p}=$"+str(proposal_sigma))
plt.show()
```

## Burn-in

We can now see our problem quite clearly. Recall that MCMC constructs a Markov chain which has stationary distribution that coincides with our target distribution. What is happening here is that we have initialised our chain at $-10$ and it takes a short while until it reaches the area of space which has large probability density and has reached the stationary distribution of the chain.

These early samples incur *starting biases* which results in our estimators for the mean and variance having some bias. Therefore, in practice, on often discards an initial set of samples which are referred to as *burn-in*.

Continuing with this example, lets choose to cut off the first $2000$ iterations of our chain...

```python
burn_in = 2000
plt.plot(range(1, len(sampler.chain)+1), sampler.chain)
plt.vlines(x=burn_in, ymin=min(sampler.chain), ymax=max(sampler.chain), colors='orange')
plt.xlabel("iteration")
plt.ylabel(r"$\theta$")
plt.title(r"Trace plot of RWMH sampler with $\sigma_{p}=$"+str(proposal_sigma))
plt.legend(['RWMH chain', 'Arbitrary cut-off'])
plt.show()
```

```python
plt.hist(sampler.chain[burn_in+1:], bins = 100, density=True)
x = np.linspace(mu-4*sigma, mu+4*sigma, 1000)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.xlabel(r"$\theta$")
plt.ylabel("density")
plt.ylim([0, 0.7])
plt.legend(['True target density', 'Histogram approximation with N={}'.format(num_samples-burn_in)])
plt.title(r"RWMH with $\sigma_{p}$="+"{} (after removing burn-in)".format(proposal_sigma))
print("Mean of MCMC output (after removing burn-in):", np.mean(sampler.chain[burn_in+1:]))
print("Variance of MCMC output (after removing burn-in):", np.var(sampler.chain[burn_in+1:]))
```

If we now extend our chain so that we can have a sample size of $100000$, we can see that we get a better approximation to the true target density and our Monte Carlo estimates of the mean and variance are closer to the true values too!

```python
extend_num_samples = 100000
sampler.MH_sample(extend_num_samples-num_samples+burn_in)
plt.hist(sampler.chain[burn_in+1:], bins = 100, density=True)
x = np.linspace(mu-4*sigma, mu+4*sigma, 1000)
plt.plot(x, norm.pdf(x, mu, sigma))
plt.xlabel(r"$\theta$")
plt.ylabel("density")
plt.ylim([0, 0.7])
plt.legend(['True target density', 'Histogram approximation with N={}'.format(extend_num_samples)])
plt.title(r"RWMH with $\sigma_{p}$="+"{} (after removing burn-in)".format(proposal_sigma))
print("Mean of MCMC output (after removing burn-in):", np.mean(sampler.chain[burn_in+1:]))
print("Variance of MCMC output (after removing burn-in):", np.var(sampler.chain[burn_in+1:]))
```

## Choosing the 'right' proposal

Here in our very simple example above, we used a Normal distribution centred at the *current* location of the chain with variance $\sigma_{p}^{2}=0.5^{2}$. We can consider how our algorithm changes by using different choices for $\sigma_{p}^{2}$.

```python
test_sigmas = [0.01, 0.1, 1, 10, 100, 1000]
plt.subplots(3, 2, figsize=(12,8))
plt.subplots_adjust(wspace = 0.2, hspace = 0.5)
for i in range(0, len(test_sigmas)):
    MH_sampler = MetropolisHastings(start_point,
                                    gen_proposal(test_sigmas[i]),
                                    log_proposal(test_sigmas[i]),
                                    log_posterior(mu, sigma))
    MH_sampler.MH_sample(10000)
    ax = plt.subplot(3, 2, i+1)
    ax.set_ylim(np.min([mu-4*sigma, mu+4*sigma, start_point]),
                np.max([mu-4*sigma, mu+4*sigma, start_point]))
    ax.set_xlabel(r"$\theta$")
    ax.set_ylabel("density")
    ax.set_title(r"$\sigma_{p}$="+str(test_sigmas[i]))
    ax.plot(range(1, len(MH_sampler.chain)+1), MH_sampler.chain)
    ax.vlines(x=burn_in,
              ymin=np.min([mu-4*sigma, mu+4*sigma, start_point]),
              ymax=np.max([mu-4*sigma, mu+4*sigma, start_point]), colors='orange')
    ax.hlines(y=mu, xmin=0, xmax=10000, colors = 'red')
```

## Autocorrelation plots

If the chain is not sampling from the distribution efficiently, we typically say that its "mixing poorly". Recall that MCMC constructs a Markov chain and so by definition, the samples are *not* independent. We can plot the *autocorrelation plot* (or *ACF*, or *correlogram*), which plots the autocorrelation at lag $k$ as a function of $k$.

The wikipedia for correlogram says:
> The correlogram is a commonly used tool for checking randomness in a data set. If random, autocorrelations should be near zero for any and all time-lag separations. If non-random, then one or more of the autocorrelations will be significantly non-zero.

We can use this to diagnose how independent our samples are. Typically, we would ideally see little correlation between our samples. In this example, we can see that when the variance of our proposal is very low, we have very high correlation between our samples - this is because the samples don't have the opportunity to move very far at each iteration. When the variance of the proposal is too high, we also get high correlation. This is due to the fact that our proposals keep getting rejected (see above trace plots), and so we are typically staying the same place at each iteration. Again, we see this 'goldilocks' sitatuion where we are getting much lower correlation between samples when our proposal is appropriate and the chain mixes well.

```python
plt.subplots(3, 2, figsize=(12,12))
for i in range(0, len(test_sigmas)):
    MH_sampler = MetropolisHastings(start_point,
                                    gen_proposal(test_sigmas[i]),
                                    log_proposal(test_sigmas[i]),
                                    log_posterior(mu, sigma))
    MH_sampler.MH_sample(10000)
    ax = plt.subplot(3, 2, i+1)
    plot_acf(MH_sampler.chain, ax = ax, lags = 100, bartlett_confint = False, alpha = None,
             title = r"$\sigma_{p}$="+str(test_sigmas[i]))
```

## A note on MCMC diagnostics

The success of MCMC relies on the chain reaching stationarity (relatively quickly) and being able to explore the space efficiently. In the examples we consider in this notebook, we know exactly what we're looking for. But in real world problems, we are using MCMC to sample from our posterior *because* we don't know what it looks like. In practice, we need to decide things like how many samples we need to discard for burn in, how long we need to run the chain after to get a large enough sample size to get 'good' estimates, and what makes a good proposal. Unfortunately, in such problems, we cannot truly know if our choices were good and whether or not the sampler has explored a space properly or if our chain has reached stationarity.

However, one can apply several graphical (e.g. considering trace plots and ACF plots), statistical tests and summary statistics to roughly assess if the chain has stabilised and reached stationarity. We have seen above that trace plots give us a visual representation of the chain and for the above example, it's easy for us to see at which point would be appropriate to discard burn-in samples. 

These simple heuristics can be used and can help us build up some confidence that we are getting the correct answer. However, in general, these heuristics do not perfectly provide satisfactory diagnostics. This is the key complication for practioners with implementing MCMC.

We won't discuss MCMC diagnostics in a lot of detail in this notebook, and for the purposes of our examples, considering trace plots will be sufficient for us to get an idea of what we want from our MCMC sampler. But for reference, some common diagnostic tools practioners look at are:

* Investigate trace plots for each parameter
* Investigate autocorrelation (ACF) plots
* Split the chain up and compare summary statistics - should be similar if has reached stationarity
* Run multiple chains with different starting points to see if they converge to the same space
    * Potentially very useful if the posterior exhibits multi-modality
* Compute the Effective sample size (ESS)
    * This is computed using the autocorrelation and gives an rough estimate of how many 'equivalent i.i.d. samples' our MCMC output provides.
* Compute the Gelman-Rubin statistic
    * See [Stata blog post on Gelman-Rubin statistic](https://blog.stata.com/2016/05/26/gelman-rubin-convergence-diagnostic-using-multiple-chains/#:~:text=The%20Gelman%E2%80%93Rubin%20diagnostic%20evaluates,between%20these%20variances%20indicate%20nonconvergence.)
* Compute the Integrated autocorrelation time (related to looking ACF plots)
    * If this is too large, can *slice*, or *thin*, the chain (i.e. take every 10th sample for instance - means discarding many more samples)
* Compute Rhat statistic
    * R-hat, or the potential scale reduction factor, is a diagnostic that attempts to flag situations where the MCMC algorithm has failed converge (see [Andrew Gelman's blog on Rhat](https://statmodeling.stat.columbia.edu/2019/03/19/maybe-its-time-to-let-the-old-ways-die-or-we-broke-r-hat-so-now-we-have-to-fix-it/#:~:text=R%2Dhat%2C%20or%20the%20potential,MCMC%20algorithm%20has%20failed%20converge.)
* **Do not just rely on a single statistic/diagnostic tool**
    * no single diagnostic is perfect;
    * we cannot 100% guarantee the quality of an MCMC sample is adequate - diagnostics simply can help us to identify a potential problem but do not guarantee the absence of problems
    
Some helpful resources:
* https://www.statlect.com/fundamentals-of-statistics/Markov-Chain-Monte-Carlo-diagnostics

In [Part II - Supplementary Material](https://rchan26.github.io/pdfs/talks/bayesian_inference_talk_part_ii/part_ii_supplementary.ipynb), we will look at a applying our MCMC sampler to a bimodal example. We see that multimodality is one of the many difficulties one can face when implementing MCMC and we will see that *just* looking at the trace plots or autocorrelation plots is not sufficient in ensuring that our sampler has explored all *significant* areas of the space.

## Simple linear regression model

While it seems like it can be cumbersome implementing MCMC (well this is somewhat true), we are  fortunate to have plenty of *probabilistic programming langauges* which have implemented many of the state of art MCMC algorithms, such as
* [PyMC3](https://docs.pymc.io/en/v3/index.html)
* [Stan](https://mc-stan.org/)
* [TensorFlor Probability](https://www.tensorflow.org/probability)
* [Edward](http://edwardlib.org/)
* [Pyro](https://pyro.ai/) and [NumPyro](https://num.pyro.ai/en/latest/index.html#introductory-tutorials)
* and many more..

Most of these implement the *No-U-Turn Sampler (NUTS)* which ia *Hamiltonian Monte Carlo* algorithm by default, but other MCMC algorithms such as Metropolis-Hastings will be readily available too.

Here we revisit the linear regression model which we saw in Part I:

$$
y \sim \mathcal{N}(\mu, \sigma) \\
\mu = \alpha + \beta x
$$

where $\alpha$ is the intercept term, $\beta$ is the slope of our linear regression line, and $\sigma$ is the standard deviation which represents the residual variability in our data that we can't explain, which results in the data not falling exactly on the mean line (e.g., due to noise in measurements or due to limitations of our model). Here we will simulate some data from this model with $\alpha=0$, $\beta=2$ and $\sigma=5$.

```python
np.random.seed(SEED)
# generate n data points over a range of x values
n = 200
xs = np.linspace(5, 15, n)

# alpha = intercept, beta = slope
alpha = 0
beta = 2
true_regression_line = alpha + beta*xs

# add normally distributed noise to data
sigma = 5
noise = norm.rvs(0, sigma , n)
ys = true_regression_line + noise

plt.scatter(xs, ys, marker="x", label="data", alpha=0.8)
plt.plot(xs, true_regression_line, color='red', label="true regression line")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

We now would like to find the values of the model parameters $\theta=(\alpha, \beta, \sigma)$ that fits our data well. Whilst in Part I, we assumed that know the residual variability $\sigma$, we will perform inference for $\sigma$ here too.

Recalling from Part I, the likelihood in this example is simply the normal distribution with mean $\mu=\alpha+\beta x$ and variance $\sigma^{2}$. For the prior distributions, we propose:

$$
\begin{align}
    \alpha & \sim \mathcal{N}(0,1) \\
    \beta & \sim \mathcal{N}(0,1) \\
    \sigma & \sim \text{Half-Cauchy}(10)
\end{align}
$$

```python
fig, ax = plt.subplots(1,2, figsize=(12,4))
x = np.arange(-5, 5, 0.01)
sigmas = np.arange(0, 50, 0.01)
ax[0].plot(x, norm.pdf(x, 0, 1))
ax[0].set_xlabel(r"$\alpha,\beta$")
ax[0].set_ylabel("density")
ax[0].set_title(r"Prior for $\alpha, \beta$")
ax[1].plot(sigmas, halfcauchy.pdf(sigmas, scale=10))
ax[1].set_xlabel(r"$\sigma$")
ax[1].set_title(r"Prior for $\sigma$")
plt.show()
```

We can implement the Bayesian model as follows by defining clearly what the prior distribution and likelihood distributions are.

Note that many of these probabilistic programming languages are able to utilise a number of cores and run a number of parallel chains - this can help with saving computation as well as diagnosing the convergence of our chains! Here, we run $4$ parallel chains.

```python
with pm.Model() as model:  # model specifications in PyMC3 are wrapped in a with-statement
    # Define priors
    sigma = pm.HalfCauchy("sigma", beta=10)
    alpha = pm.Normal("alpha", 0, sigma=1)
    beta = pm.Normal("beta", 0, sigma=1)

    # Define likelihood
    likelihood = pm.Normal("y", mu=alpha + beta * xs, sigma=sigma, observed=ys)

    # Inference!
    # draw 3000 posterior samples using NUTS sampling
    trace = pm.sample(2500, tune=2000, chains=4, return_inferencedata=True, random_seed=SEED)
```

We can use the `arviz` package to look at trace and autocorrelation plots:

```python
az.plot_trace(trace)
```

```python
az.plot_autocorr(trace, combined=True)
```

From the trace and autocorrelation plots, it looks like the sampler was very efficient! By inspecting the posterior distributions for each of the chains, we can see that that our sampler seems to have concentrated around the true values of the parameters.

Note that by default, `pymc3` plots out the autocorrelation plots for each chain, but we can combined them for a simpler plot.

We can also obtain summary statistics easily too:

```python
az.summary(trace)
```

Looks like the means for each of the parameters is close to the true values which were used to simulate the data: $(\alpha, \beta, \sigma)=(0,2,5)$!

We can use our samples to obtain a *posterior predictive distribution* for new data. For instance, if we had a new data point $x=20$, we can use each of our samples $\{\alpha_{i}, \beta_{i}, \sigma_{i}\}_{i=1}^{N}$ to generate predictions from our likelihood:

$$
y_{i} \sim \mathcal{N}(\alpha_{i} + \beta_{i} x, \sigma_{i}).
$$

for $i=1,\dots,N$.

```python
# obtain posterior samples for each parameter
posterior_alpha = trace.posterior.alpha.data.reshape(10000)
posterior_beta = trace.posterior.beta.data.reshape(10000)
posterior_sigma = trace.posterior.sigma.data.reshape(10000)

# generate new data using posterior samples
x_star = 20
y_star = posterior_alpha + posterior_beta*x_star + norm.rvs(0, posterior_sigma)

plt.hist(y_star, density=True, bins=100)
plt.xlabel(r"$y$")
plt.ylabel("density")
plt.title(fr"Posterior predictive for $x={x_star}$")
plt.show()
```

We can summarise this by plotting each of the lines that we get from our samples:

```python
for i in range(0, len(posterior_alpha)):
    # plot regression line generated with ith sample
    plt.plot(xs, posterior_alpha[i] + posterior_beta[i]*xs + norm.rvs(0, posterior_sigma[i]),
             c='gray', alpha=0.01);
plt.plot(xs, posterior_alpha.mean() + posterior_beta.mean() * xs, c='k',
         label='y = {:.2f} + {:.2f}* x'.format(posterior_alpha.mean(), posterior_beta.mean()))
plt.plot(xs, true_regression_line, color='red', label="true regression line")
plt.scatter(xs, ys, marker="x", label="data", alpha=1)
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()
```

Note that there are ready built functions to generate these type of plots, e.g. `plot_hdi` and `plot_ppc` from the [`arviz`](https://arviz-devs.github.io/arviz/) package!

## Conclusions

* In many cases of Bayesian inference, we must have some tool to approximate the posterior distribution
    * This is to avoid having to evaluate the intractable evidence/normalising constant in the Bayes formula
* MCMC sampling is an efficient and flexible mechanism for sampling from the posterior distribution
* By obtaining samples from the posterior, we can obtain quantities related to it easily, i.e. we can perform predictions, obtain summary statistics, make statements like "I believe the parameter is in the interval..." (credible intervals), among other things, with little effort
